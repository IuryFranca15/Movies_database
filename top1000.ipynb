```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')
```

```python
df.info()
```

```python
df = df.drop(columns= 'Poster_Link')
```

```python
#quais são as produções com maior nota
filtro = df[['Series_Title', 'IMDB_Rating']].sort_values(by= 'IMDB_Rating', ascending= False)
filtro.head(10)
```

```python
#quais são os filmes de Scorsese na lista?
filtro = df[df['Director'] == 'Martin Scorsese'][['Series_Title', 'IMDB_Rating']]
filtro.sort_values(by= 'IMDB_Rating', ascending= False)
```

```python
#quais são os generos que mais aparecem na lista?
filtro = df.groupby('Genre')['Series_Title'].count()
filtro.sort_values(ascending=False).head()
```

```python
#ou
top_generos = df['Genre'].value_counts().head()
print(top_generos)
```

```python
#quais são os diretores com mais filmes na lista?
top_diretores = df['Director'].value_counts().head()
print(top_diretores)
```

```python
#quais são os filmes mais bem avaliados dos anos 80?
filtro80 = (df['Released_Year'] >= 1980) & (df['Released_Year'] <= 1989)
df80 = df[filtro80]
top10 = df80[['Series_Title', 'IMDB_Rating']].sort_values(by= 'IMDB_Rating', ascending = False).head(10)
print(top10)
```

```python
#quais são os filmes mais bem avaliados dos anos 90?
filtro90 = (df['Released_Year'] >= 1990) & (df['Released_Year'] <= 1999)
df90 = df[filtro90]
top10 = df90[['Series_Title','IMDB_Rating']].sort_values(by='IMDB_Rating', ascending= False).head(10)
print(top10.sort_index())
```

```python
#e dos anos 2000?
filtro2000 = (df['Released_Year'] >= 2000) & (df['Released_Year'] <= 2009)
df2000 = df[filtro2000]
top10 = df2000[['Series_Title','IMDB_Rating']].sort_values(by='IMDB_Rating', ascending= False).head(10)
print(top10)
```

```python
#quais são os 5 diretores que mais faturaram em média, mas apenas
#considerando filmes que tenham uma nota do IMDB maior que 8.0
filtro = df[df['IMDB_Rating'] > 8.0][['Director', 'Gross']]
filtro['Gross'] = filtro['Gross'].str.replace(",", "").astype(float)
top_diretores = filtro.groupby('Director')['Gross'].mean().sort_values(ascending=False).head()
print(top_diretores)
```

```python
#limpando o 'min' da celula e transformando em
df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(int)
```

```python
#Qual é a média de nota (IMDB_Rating) dos 
#filmes que duram mais de 2 horas (120 min)

filtro = df[df['Runtime'] > 120]
media = filtro['IMDB_Rating'].mean()
print(f'A media é {media:.2f}')
```

```python
#quais os generos mais lucrativos
df['Gross'] = pd.to_numeric(df['Gross'].str.replace(',', ''), errors='coerce')

ranking_genero = df.groupby('Genre')['Gross'].sum().sort_values(ascending=False)

print("Os 3 Gêneros mais lucrativos (em milhões):")
print(ranking_genero.head(3) / 1_000_000)
```

```python
#quais são os 10 diretores com maior media de nota de filmes?
medias_por_diretor = df.groupby('Director')['IMDB_Rating'].mean()
medias_por_diretor = medias_por_diretor.sort_values(ascending=False).head(10)
print(medias_por_diretor)
```

```python
# e quais são os filmes do diretor com maior media?
diretor = df[df['Director'] == 'Frank Darabont']
filmes = diretor[['Series_Title', 'IMDB_Rating']].sort_values(by= 'IMDB_Rating', ascending = False)
print(filmes)
```
